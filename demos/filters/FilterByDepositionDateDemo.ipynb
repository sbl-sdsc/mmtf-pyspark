# Filter By Deposition Date Demo

## Imports

```python
from pyspark import SparkConf, SparkContext
from mmtfPyspark.io import MmtfReader
from mmtfPyspark.filters import depositionDate
```

## Configure Spark

```python
conf = SparkConf().setMaster("local[*]") \
                      .setAppName("FilterByDepositionDate")
sc = SparkContext(conf = conf)
```

## Read in MMTF Files

```python
path = "full path to your MMTF file directory"

pdb = MmtfReader.readSequenceFile(path, sc)
```

## Filter by deposition date

```python
pdb = pdb.filter(depositionDate('2016-01-28','2016-02-28'))
```

## Count number of entires

```python
count = pdb.count()

print(f"Number of structure desposited between 2016-01-28 and 2017-02-28 is : {count}")
```

## Terminate Spark 

```python
sc.stop()
```
